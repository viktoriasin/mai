**Задание**: Разработать и реализовать программу для работы с изображениями фишек игрового набора
Пазл (Puzzle20), обеспечивающую:
* Ввод и отображение на экране изображений;
* Сегментацию изображений на основе точечных и пространственных
преобразований;
* Выделение отдельных фишек для последующей классификации;
* Подсчет числа фишек на изображении.

**Входные данные**: в качестве входных данных используются изображения
игровых фишек двух типов. Изображения первого типа содержат почти
однордный задний план и лицевую сторону фишек, второго - пестрый фон и заднюю сторону фишек.

**Метод решения**: Для решения задачи используются два алгоритма в
зависимости от типа изображений. Для изображений первого типа применяется
функция `invert_red` (см. исходный код ниже), которая заменяет красный фон
заднего плана на черный, после чего оставшаяся часть изображения
заливается белым цветом.

Для изображений второго типа алгоритм отиличается только тем, что теперь
однотонные фишки отделяются от остального фона.

После успешного выделения фишек применяется алгоритм нахождения контуров.
Принцип работы алгоритма подробно изложен в [2]. Для выполнения данной лабораторной работы использовалась реализация из библиотеки OpenCV.
После нахождения контуров необходимо отделить контуры дефектов от
контуров фишек. Для этого контуры сортируются по значению площади,
после чего контуры с площадью меньше некоторого значения $S_0$
отбрасываются. Значение $S_0$ было найдено эмпирическими методами.

**Описание программной реализации**:
Программа написана на языке программирования Python3 с использованием
библиотек Matplotlib, Numpy, Scikit-Image и OpenCV

**Литература**:
1. Гонзалес Р., Вудс Р. Цифровая обработка изображений. М., Техносфера, 2006.
2. Satoshi Suzuki and others. Topological structural analysis of digitized binary images by border following. Computer Vision, Graphics, and Image Processing, 30(1):32–46, 1985.

```python
import skimage as si
import skimage.io
import skimage.morphology
import skimage.exposure

import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

import os
```

```python
def invert_red(image):
    red = 150
    other = 110
    image = np.copy(image)
    image[(image[:, :, 0] > red) & (image[:, :, 1] < other) & (image[:, :, 2] < other)] = 0
    image[image != 0] = 255
    
    image = skimage.color.rgb2gray(image)
    image[image > 0.6] = 255
    image[image <= 0.6] = 0
    
    image = si.morphology.area_opening(image, area_threshold=512)
    image = si.morphology.binary_dilation(image)
    
    return si.util.invert(image).astype(np.uint8)
```

### Подсчет пазлов на однотонном красном фоне

```python
images = ['Red_3.jpg', 'Red_4.jpg']
images = [os.path.join('images', i) for i in images]

for i in images:
    image = si.io.imread(i)
    proc_image = invert_red(image)
    
    contours, hierarchy = cv.findContours(proc_image.copy(), cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
    #sort_contours = sorted(contours, reverse=True, key=lambda c: cv.contourArea(c))
    filtered_contours = list(filter(lambda c: cv.contourArea(c) > 50000, contours[1:]))

    print('Count: ', len(filtered_contours), ',', [cv.contourArea(c) for c in filtered_contours])
    
    for c in filtered_contours:
        rect = cv.minAreaRect(c)
        box = cv.cv2.boxPoints(rect)
        box = np.int0(box)
        cv.drawContours(proc_image,[box],0,(0,0,255),2)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(16, 9))
    ax1.imshow(image)
    ax2.imshow(proc_image)
    plt.show()
```

### Подсчет пазлов на пестром фоне

```python
def invert_blue_puzzle(image):
    color = 120
    other = 230
    image = np.copy(image)
    
    p1, p2 = np.percentile(image, (5, 95))
    image = si.exposure.rescale_intensity(image, in_range=(p1, p2))
    image[(image[:, :, 2] > color) & (image[:, :, 0] < other) & (image[:, :, 1] < other)] = 0
    image[image != 0] = 255
    
    image = skimage.color.rgb2gray(image)
    image[image > 0.6] = 255
    image[image <= 0.6] = 0
    
    image = si.morphology.area_opening(image, area_threshold=512)
    image = si.morphology.binary_dilation(image)
    
    return image.astype(np.uint8)
```

```python
images = [f'Motley-{i}.jpg' for i in range(12, 18)]
images = [os.path.join('images', i) for i in images]

for i in images:
    image = si.io.imread(i)
    proc_image = invert_blue_puzzle(image)
    
    contours, hierarchy = cv.findContours(proc_image.copy(), cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
    sort_contours = sorted(contours, reverse=True, key=lambda c: cv.contourArea(c))
    filtered_contours = list(filter(lambda c: cv.contourArea(c) > 5000, sort_contours[1:]))
    print('Count: ', len(filtered_contours), ',', [cv.contourArea(c) for c in filtered_contours])
    
    for c in filtered_contours:
        rect = cv.minAreaRect(c)
        box = cv.cv2.boxPoints(rect)
        box = np.int0(box)
        cv.drawContours(proc_image,[box],0,(0,0,255),2)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(16, 9))
    ax1.imshow(image)
    ax2.imshow(proc_image)
    plt.show()
```
