```python
%matplotlib inline
import numpy as np
import cv2
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.preprocessing import StandardScaler, PowerTransformer
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = [15, 15]
```

```python
def angle_rad(v1, v2):
    return np.arctan2(np.linalg.norm(np.cross(v1, v2)), np.dot(v1, v2))
```

```python
def chi2_distance(histA, histB, eps = 1e-10):
    # мера расстояния chi-квадрат
    d = 0.5 * np.sum([((a - b) ** 2) # / (a + b + eps)
                      for (a, b) in zip(histA, histB)])
    return d
```

```python
def rank(query_features, index_features):
    # вычисляем расстояние до изображения из базы
    result = np.zeros(len(index_features))
    for i in range(len(index_features)):
        result[i] = chi2_distance(query_features, index_features[i])
    return result
```

```python
class HandFeauturesDescriptor:
    
    def mark_hand_center(self, frame_in, cont, draw=False, radius_thresh=0.04):    
        max_d=0 #Максимальное расстояние от контура до точки
        pt=(0,0) #Начальное приближение точки центра ладони
        x,y,w,h = cv2.boundingRect(cont) #Считает координаты минимального прямоугольника вокруг контура
        for ind_y in range(int(y+0.3*h),int(y+0.8*h)): # Достаточно проходить только по этому интервалу, чтобы найти центр
            for ind_x in range(int(x+0.3*w),int(x+0.6*w)): 
                dist= cv2.pointPolygonTest(cont,(ind_x,ind_y),True) # Ищет расстояние от котура до точки
                if(dist>max_d):
                    max_d=dist
                    pt=(ind_x,ind_y)
        if(max_d>radius_thresh*frame_in.shape[1]):
            thresh_score=True
            if draw:
                cv2.circle(frame_in,pt,1,(255,0,0),2)
        else:
            thresh_score=False
        return pt
    
    def describe(self,path, img, center_half=300, kernel_size = 5, draw=True):
    
        base_size = 9 # количество признаков в массиве
        
        # переводим в серую цветовую шкалу
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        # пытаемся выделить регион руки (предполагаем что рука расположена примерно по центру)
        height, width = gray.shape
        center_half = center_half
        center = gray[height // 2 - center_half:height // 2 + center_half,
                       width // 2 - center_half:width // 2 + center_half]
        med_val = np.median(center)
        #все темные области делаем черным, остальное серым
        frame = np.where(abs(gray - med_val) <= 14,
                         0, 128).astype(np.uint8)
        #пытаемся избавиться от мелких дыр с помощью операции закрытия
        kernel = np.ones((kernel_size, kernel_size), np.uint8)
        frame = cv2.morphologyEx(frame, cv2.MORPH_CLOSE, kernel)
        
        # заполняем область руки, чтобы окончательно избавиться от дыр и разрывов
        small_kernel = 3
        # помечаем начальную точку для алгоритма cv2.floodFill
        frame[height // 2 - small_kernel:height // 2 + small_kernel,
          width // 2 - small_kernel:width // 2 + small_kernel] = 128
        mask = np.zeros((height + 2, width + 2), np.uint8)
        flood = frame.copy()
        cv2.floodFill(flood, mask, (width // 2, height // 2), 255,
                      flags=4 | (255 << 8))
        ret, flooded = cv2.threshold(flood, 129, 255, cv2.THRESH_BINARY)
        #дополнительно уменьшаем изображение чтобы избавиться от мелких ненужных пятен на изображении
        img_erode = cv2.erode(flooded,kernel, iterations=5)
        # ищем контуры на изображении
        contours, hierarchy = cv2.findContours(img_erode, cv2.RETR_TREE,
                                               cv2.CHAIN_APPROX_SIMPLE)
        if not contours:
            return np.zeros(base_size)
        # считаем, что самый большой по площади контур - это рука
        max_contour = max(contours, key=cv2.contourArea)
        max_contour_for_center = max_contour.copy()
        # оставляем только необходимые точки для контура, чтобы проще было потом находить дефекты
        epsilon = 0.015 * cv2.arcLength(max_contour, True)
        max_contour = cv2.approxPolyDP(max_contour, epsilon, True)
        med = np.median(max_contour,axis=0)
        # строим выпуклую оболочку контура
        hull = cv2.convexHull(max_contour, returnPoints=False)
        # ищем участки оболочки которые лежат между пальцами
        defects = cv2.convexityDefects(max_contour, hull)
        
        img_draw = img.copy()
        
        # ищем координаты центра
        cnter_coordinates = self.mark_hand_center(img_draw, max_contour_for_center, draw=False)
        
        
        features = []
        if defects is None:
            return np.zeros(base_size)
        for defect in defects[:, 0, :]:

                start, end, far = [max_contour[i][0] for i in defect[:3]]
                if end[1] > med[0][1] * 1.3  or start[1] > med[0][1] * 1.3:
                    continue
                
                #angle between fingers
                angle_fing = angle_rad(start - far, end - far)

                #angle between center and fingers
                angle_betw = angle_rad(start - cnter_coordinates, end - cnter_coordinates)
                #distance between center and finger
                dist_fing = euclidean_distances([cnter_coordinates,tuple(start)])[0][1]
                #distance between center and area between fingers
                dist_betw = euclidean_distances([cnter_coordinates,tuple(far)])[0][1]
                
                features.extend([dist_fing,dist_betw]) #angle_fing,angle_betw,

                if draw:
                    cv2.line(img_draw, tuple(start), tuple(far), (0, 255, 0), 2)
                    cv2.line(img_draw, tuple(far), tuple(end), (0, 255, 0), 2)

                    cv2.line(img_draw, tuple(start), tuple(start), (255, 0, 155), 10)
                    cv2.line(img_draw, tuple(end), tuple(end),(255, 0, 155), 10)
                    cv2.line(img_draw, tuple(far), tuple(far), (255, 0, 155), 10)
        if draw:
            plt.figure()
            plt.title(path);
            plt.imshow(cv2.cvtColor(img_draw, cv2.COLOR_BGR2RGB))

        #отдельно считаем расстояние до большого пальца
        dist_fing = euclidean_distances([cnter_coordinates,tuple(end)])[0][1]
        features.append(dist_fing)

        #print(len(features))
        if len(features) != base_size:
            return np.zeros(base_size)
        
        return features
```

```python
import glob

index = []   # имена фалов
index_features = []  # дескрипторы изображений
hand_features = []

# получаем список файлов для индексации
files = list(glob.glob('./images/training/*.tif'))

# инициализируем объект для вычисления дескрипторов
cd = ColorDescriptor((8, 12, 3))
hd = HandFeauturesDescriptor()

for path in files:
    # сохраняем идентификатор изображения в массив
    image_id = path[path.rfind('/') + 1:]
    index.append(image_id)

    # загружаем изображение
    image = cv2.imread(path)
    
    # вычисляем и сохраняем дескрипторы
    hd_features = hd.describe(path, image, draw=False)
    hand_features.append(hd_features)

hand_features = np.array(hand_features)
```

```python
# загружаем изображение для поиска похожих
query = cv2.imread('./images/training/test/146.tif')
# 146 001 078
# получаем фичи для изображения запроса
query_features_h = np.array(hd.describe(query)).reshape(-1,1)
plt.title('Запрос');

# ранжируем резльтат с учетом метрики расстояния
distance = rank(query_features_h, hand_features)

# сортируем индексы по расстоянию
indices = np.argsort(distance)
```
